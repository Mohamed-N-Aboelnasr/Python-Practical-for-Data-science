# **Discrete Fourier Transform**

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
fs=32000
ts=1/fs
l=1024
t=np.arange(l)*ts
fc=0.25*(fs/2)
x=np.sin(2*np.pi*t*fc)
fc=0.5*(fs/2)
x+=0.5*np.sin(2*np.pi*t*fc)
plt.plot(t,x)
```

```python
#replace the pass with your code
def myDFT(x):
    #implement the DFT Euation
    #https://en.wikipedia.org/wiki/Discrete_Fourier_transform
    #use the equation from the definition section
    
    # for loop sol
    #import numpy as np
    #N = len(x)
    #i = np.complex(0,1)
    #Y = np.empty((x.shape))
    #Y.fill(0)
    #for k in range(N):
    #  for n in range(N):
        #Y[k] += np.exp(-2*i*np.pi*k*n/N) * x[n]
    #return Y
    
    
    # numpy sol
    n=np.arange(len(x))
    k=n.reshape(len(x),1)
    exp_term=np.exp(-2j*np.pi*k*n/len(x))
    x_k=np.dot(x,exp_term)
    return x_k

def DFT_shift(x):
    #shift rotate the victor for better visialiation
    #i/p x[0:n]
    #o/p x[n/2 : n, 0:n]

    n = len(x)
    return np.concatenate((x[int(n/2):n], x[0:int(n/2)]), axis=None)
```

```python
y=myDFT(x)
f=np.linspace(-1,1,len(y))
plt.plot(f,DFT_shift(np.abs(y)))
```
