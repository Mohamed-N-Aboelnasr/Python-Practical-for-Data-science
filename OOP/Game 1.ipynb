# Game - Object-Oriented Programming Task 1

```python
import math

class moving_obj():
    def __init__(self):
        self.v=0
        self.vdirction=0 
        #0---->+y || 1---->+x || 2---->-y || 3---->-x
        self.x=0
        self.y=0
    
    def update(self, command):
        #location update
        self.x+=(self.vdirction==1)*self.v-(self.vdirction==3)*self.v
        self.y+=(self.vdirction==0)*self.v-(self.vdirction==2)*self.v
        
        #velocity update
        if command=='w':
            self.v+=1
        elif command=='s':
            self.v-=1
        elif command=='a':
            self.vdirction=(self.vdirction-1)%4
        elif command=='d':
            self.vdirction=(self.vdirction+1)%4                    
    
    def display(self):
        dir_dict={0:'+y', 1:'+x' , 2:'-y',  3:'-x'}
        print('v =',(self.v,dir_dict[self.vdirction]))
        print('pos=',(self.x, self.y))

class player(moving_obj):
    def __init__(self,name):
        self.name=name
        super().__init__()
    #target=(x,y)
    def iswon(self,target):
        if ((self.x,self.y)==target):
            return True
        else:
            return False 

class enemy(moving_obj):
    def __init__(self,name,target):
        self.name=name
        super().__init__()
        self.x=target[0]
        self.y=target[1]
    #target=(x,y)
    def iswon(self,player):
        if math.sqrt((self.x-player.x)**2+(self.y-player.y)**2)<=1:
            return True
        else:
            return False
        
# Display the distance between players and enemies
def display_dist(players,enemies):
    for player in players:
        for enemy in enemies:
            print("Distance between (player)"+player.name+" and (enemy)"+enemy.name+" : ",math.sqrt((player.x-enemy.x)**2+(player.y-enemy.y)**2))

#####################################  Main ########################################
target=(int(input('target_x= ')), int(input('target_y= ')))
n=int(input('Enter number of players : '))
n2=int(input('Enter number of enemies : '))
max_c=int(input('Enter number of steps you want the game to exit at : '))
players=[player(input('Enter player name :')) for i in range(n)]
enemies=[enemy(input('Enter enemy name :'),target) for i in range(n)]
counter=0
while True:
    # General display
    counter+=1
    print('=================== target',target,'=================== w,s,d,a,end')
    leave=0
    index=[]
    # Read and update player
    for i in range(len(players)):
        command=input("(player) "+players[i].name+' ')
        if command=='end':
            print('(player) '+players[i].name+' left')
            leave=1
            index.append(i)
        else:
            players[i].update(command)
        print('-------------')
    if leave==1:
        c=0
        for i in index:
            del(players[i-c])
            c+=1
    leave=0
    index=[]
    # Read and update enemy
    for i in range(len(enemies)):
        command=input("(enemy) "+enemies[i].name+' ')
        if command=='end':
            print('(enemy) '+enemies[i].name+' left')
            leave=1
            index.append(i)
        else:
            enemies[i].update(command)
        print('-------------')
    # Display players status
    for i in range(len(players)):
        print('(player) '+players[i].name)
        players[i].display()
        print("----------------")
    # Display enemies status
    for i in range(len(enemies)):
        print('(enemy) '+enemies[i].name)
        enemies[i].display()
        print("----------------")
    # Display distance between players and enemies
    ans=input("Do you want to show the distance between players and enemies : [yes] , [no] ")
    if ans=="yes":
        display_dist(players,enemies)
    # Show winners
    exit=0
    for i in range(len(players)):
        if players[i].iswon(target):
            print('(player)'+players[i].name,'won!, congratulations!')
            print('number of steps = ',counter)
            exit=1
    for i in range(len(enemies)):
        for j in range(len(players)):
            if enemies[i].iswon(players[j]):
                print('(enemy)'+enemies[i].name,'won!, congratulations!')
                print('number of steps = ',counter)
                exit=1
                continue
    # Exit Game
    if exit==1 or len(players)==0 or counter==max_c:
        print('Exit the game')
        break       
    
```
